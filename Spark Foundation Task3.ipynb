```python
plt.figure(figsize=(20,15))
sns.countplot('Sub-Category',data=dataset,hue='Category')
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x=dataset.State,order=(dataset.State.value_counts().head(10)).index)
```

```python
sns.scatterplot(x='Category',y='Sales',data=dataset)
```

```python
sns.scatterplot(x='Category',y='Profit',data=dataset,color='red')
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(x='Sub-Category',y='Sales',data=dataset)
```

```python
plt.figure(figsize=(18,7))
sns.lineplot(x='Sub-Category',y='Quantity',data=dataset,legend='brief')
```

```python
state=dataset.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
state[:20].plot(kind='bar',color=['green','blue'])
plt.title('Profit and Sales in top 20 states')
plt.xlabel('States')
plt.ylabel('Total profit/sales')
```

```python
sns.barplot(y='Discount',x='Category',data=dataset)
```
